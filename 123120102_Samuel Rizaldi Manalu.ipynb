```python
# Control Flow Statements (Perulangan)
# for s in range (angka , batas, jarak)
for s in range(0,100,1):
  if s % 2 == 0:
    print(f"Bil {s} adalah genap")
  elif s % 2 != 0:
    print(f"Bil {s} adalah ganjil")
  else:
    print(f"Bil {s} adalah bukan bilangan bulat")
    break 
```

```python
# Challenge 2 : menghitung rata-rata usia keluarga
anggota = int(input("Masukkan jumlah anggota keluarga: "))
jumlah_usia = 0

for s in range(anggota):
    umur = int(input(f"Input usia anggota ke-{s+1}: "))
    jumlah_usia += umur

rata_rata = jumlah_usia / anggota
print("Rata-rata usia keluarga adalah:", round(rata_rata, 2))
```
